```python
board = [' ' for x in range(10)]
```

```python
def insertletter(letter,pos):
    board[pos] = letter
```

```python
def spaceisfree(pos):
    return board[pos]==' '
```

```python
def printboard(board):
    print('   |     |')
    print('  '+board[1]+'|  '+board[2]+'  |  '+board[3])
    print('   |     |')
    print('----------------------')
    print('   |     |')
    print('  '+board[4]+'|  '+board[5]+'  |  '+board[6])
    print('   |     |')
    print('----------------------')
    print('  '+board[7]+'|  '+board[8]+'  |  '+board[9])
    print('   |     |')
```

```python
def iswinner(bo,le):
    return (bo[7]==le and bo[8]==le and bo[9]==le) or (bo[4]==le and bo[5]==le and bo[6]==le) or (bo[1]==le and bo[2]==le and bo[3]==le) or (bo[1]==le and bo[4]==le and bo[7]==le) or (bo[2]==le and bo[5]==le and bo[8]==le) or (bo[3]==le and bo[6]==le and bo[9]==le) or (bo[1]==le and bo[5]==le and bo[9]==le) or (bo[3]==le and bo[5]==le and bo[7]==le)
```

```python
def playermove():
    run = True
    while run:
        move = input("pls enter a position to place an \'X\'(1-9):")
        try:
            move = int(move)
            if move>0 and move<10:
                if spaceisfree(move):
                    run=False
                    insertletter('X',move)
                else:
                    print("srry this place is occupied")
            else:
                print("pls enter no within range")
        except:
            print("pls type a number")
```

```python
def compmove():
    possiblemoves = [x for x, letter in enumerate(board) if letter==' ' and x!=0]
    move = 0
    
    for let in ['O','X']:
        for i in possiblemoves:
            boardcopy = board[:]
            boardcopy[i] = let
            if iswinner(boardcopy,let):
                move = i
                return move
    corneropen = []    
    for i in possiblemoves:
        if i in [1,3,7,9]:
            corneropen.append(i)
    if len(corneropen)>0:
        move = selectrandom(corneropen)
        return move
    if 5 in possiblemoves:
        move = 5
        return move
    edgesopen = []    
    for i in possiblemoves:
        if i in [1,3,7,9]:
            edgesopen.append(i)
    if len(edgesopen)>0:
        move = selectRandom(edgesopen)
    return move
```

```python
def selectrandom(li):
    import random
    ln = len(li)
    r = random.randrange(0,ln)
    return li[r]
```

```python
def isboardfull(board):
    if board.count(' ')>1:
        return False
    else:
        return True
```

```python
def main():
    print("welcome to tic tac toe")
    printboard(board)
    
    while not(isboardfull(board)):
        if not (iswinner(board,'O')):
            playermove()
            printboard(board)
        else:
            print("\n sorry o has won")
            break
        if not (iswinner(board,'X')):
            move = compmove()
            if move == 0:
                print('Tie game')
            else:
                insertletter('O',move)
                print('Computer placed an O in postion ',move," :")
                printboard(board)
        else:
            print("\n X has won good job")
            break
    if isboardfull(board):
        print("Tie game")
main()
```
